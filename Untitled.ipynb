```python
import numpy as np
import noise
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
shape = (N, N)

world_depth = 4000

height_map = np.zeros(shape)
temp_map = np.zeros(shape)
wind_map = np.zeros(shape)
humidity_map = np.zeros(shape)
cloud_map = np.zeros(shape)
rain_map = np.zeros(shape)
```

```python
# Make terrain map
scale = 100
octaves = 12
persistence = 0.6
lacunarity=2.0

grid = np.zeros(shape)
for i in range(N):
    for j in range(N):
        height_map[i][j] = noise.pnoise2(i/scale, j/scale, 
                                  octaves=octaves,
                                  persistence=persistence,
                                  lacunarity=lacunarity,
                                   repeatx=N,
                                   repeaty=N,
                                   base=0)
height_map += abs(height_map.min())
height_map *= 4000
plt.imshow(height_map)
```

```python
# Wind map
wind_direction = np.random.normal(size=2)

for x in range(N):
    for y in range(N):
        x1 = int(i + 10*(wind_direction[0]))
        y1 = int(i + 10*(wind_direction[1]))
        if (x1 < 0 or x1 > 100): x1 = x
        if (y1 < 0 or y1 > 100): y1 = y
        wind_map[x][y] = 5 * (1 - (height_map[x][y] - height_map[x1][y1])/1000)
        
plt.imshow(wind_map)
```

```python
# Temperature map
temp_map[:] = 0.7
for x in range(N):
    for y in range(N):
        if height_map[x][y] > 200:
            temp_map[x][y] = 1 - height_map[x][y]/2000
plt.imshow(temp_map)
```

```python
def make_wind_map(scale=1, u_param=1, v_param=-1):
    scale = 1
    x = np.linspace(-scale, scale, 100)
    y = x
    X, Y = np.meshgrid(x, y)
    U = u_param - X**2 + Y
    V = v_param + X - Y**2
    #speed = np.sqrt(U*U + V*V)
    return (U, V)
```

```python
plt.streamplot(x, y, *make_wind_map(1, -0.25, 0.75))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes = axes.flatten()
params = [(0, 1), (1, 0), (-1, 0), (0, -1)]
for i in range(4):
    ax = axes[i]
    u, v = params[i]
    ax.streamplot(x, y, *make_wind_map(1, u, v))
    ax.set_title((u, v))
```

```python
def rotate_vector(v, angle):
    ''' Rotate a vector by an angle
    '''
    transformation = np.array([
        [ np.cos(angle), -np.sin(angle) ],
        [ np.sin(angle),  np.cos(angle) ]
    ])
    return transformation.dot(v)
```

```python
rotate_vector(np.array([1, 0]), np.pi)
```
